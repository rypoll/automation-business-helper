## Get Links - for make

```python
chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument('--headless')  # This runs Chrome in headless mode (no GUI)
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Setup WebDriver using your methodology
driver_path = ChromeDriverManager().install()
print(f"Driver path: {driver_path}")
chrome_service = Service(executable_path=driver_path)
chrome_service.service_log_path = 'NUL'
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

all_links = []
all_types = []


for index, row in initial_df.iterrows():
    print(f"Fetching XML content from: {row['links']}")
    response = requests.get(row['links'])
    
    if response.status_code != 200:
        print(f"Failed to fetch content from: {row['links']} with status code: {response.status_code}")
        continue

    xml_content = response.text
    soup = BeautifulSoup(xml_content, 'lxml')
    links = [loc.get_text() for loc in soup.find_all('loc')]
    
    all_links.extend(links)
    all_types.extend([row['type']] * len(links))

    print(f"Extracted {len(links)} links from {row['links']}")

# Close the browser
print("Closing the browser...")
driver.quit()

# Create a new dataframe
df_new = pd.DataFrame({
    'links': all_links,
    'type': all_types
})

# Read existing links if the file exists
if os.path.exists('01a-link-list.csv'):
    print("Reading existing links from '01a-link-list.csv'...")
    df_existing = pd.read_csv('01a-link-list.csv')
    df_combined = pd.concat([df_existing, df_new]).drop_duplicates(subset=['links','content']).reset_index(drop=True)
else:
    df_combined = df_new

# Save to CSV
print("Saving combined links to '01a-link-list.csv'...")
df_combined.to_csv('01a-link-list.csv', index=False)

print("Process complete!")
```

## Get titles from Zapier

```python

# Set up Chrome options
chrome_options = ChromeOptions()
#chrome_options.add_argument('--headless')  # Run in headless mode if uncommented
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Set up the WebDriver
driver_path = ChromeDriverManager().install()
chrome_service = Service(executable_path=driver_path)
chrome_service.service_log_path = 'NUL'
driver = Chrome(service=chrome_service, options=chrome_options)

# List of URLs to scrape
urls = ['https://zapier.com/blog/all-articles/zapier-tutorials/']
for i in range(1, 15):  # From 1 to 14
    urls.append(f'https://zapier.com/blog/all-articles/zapier-tutorials/{i}')

# List to hold the scraped texts
all_texts = []

# Scrape each URL
for url in urls:
    print(f"Scraping {url}...")
    driver.get(url)
    time.sleep(10)  # Wait to ensure the page has loaded
    
    # Find all div elements with the specified data-testid attribute
    articles = WebDriverWait(driver, 30).until(
        EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div[data-testid="articleCard"]'))
    )
    
    # Extract the text from the <p> tags under each div
    for article in articles:
        p_texts = article.find_elements(By.CSS_SELECTOR, 'p')
        for p in p_texts:
            all_texts.append(p.text)
    print(f"Scraped {len(articles)} articles from {url}")

print("Finished scraping. Closing the browser...")
# Close the browser
driver.quit()

# Load the existing data (if the CSV file already exists)
existing_texts = []
csv_file = '01b-zap-blog-titles.csv'
if os.path.exists(csv_file):
    print(f"Loading existing titles from {csv_file}...")
    df_existing = pd.read_csv(csv_file)
    existing_texts = df_existing['content'].tolist()

# Combine the existing texts with the newly scraped texts and remove duplicates
all_texts = list(set(all_texts + existing_texts))

# Save the data to a CSV
print(f"Saving {len(all_texts)} unique titles to {csv_file}...")
df = pd.DataFrame(all_texts, columns=["content"])
df.to_csv(csv_file, index=False)
print("Done!")
```

## Chunk up

```python
import pandas as pd

# Load the Excel file
file_path = "01a-link-list-manycam.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Function to count tokens
def count_tokens(text):
    return len(str(text).split())

# Add a new column for token count
df['token_count'] = df['content'].apply(count_tokens)

# Function to split dataframe into chunks based on token count
def split_dataframe(df, max_tokens=3000): # use 14k
    chunks = []  # List to store chunks of the dataframe
    
    while not df.empty:
        total_tokens = 0
        rows_to_take = []
        
        for index, row in df.iterrows():
            if total_tokens + row['token_count'] <= max_tokens:
                total_tokens += row['token_count']
                rows_to_take.append(index)
            else:
                break
        
        chunk = df.loc[rows_to_take]
        chunks.append(chunk)
        
        # Remove these rows from the original dataframe
        df = df.drop(rows_to_take)
    
    return chunks

# Call the function to split the dataframe
chunks = split_dataframe(df)

# To see the chunks and verify their token counts
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1} token count: {chunk['token_count'].sum()}")
```

```python
def chunk_to_text(chunk):
    """Convert a chunk (dataframe) to the desired text format."""
    # Use list comprehension to create a list of 'source: <source_value>\n<content_value>' for each row
    lines = [f"key: {idx}\nsource: {row['source']}\n{row['content']}" for idx, row in chunk.iterrows()]
    # Join all lines into a single text separated by two newlines
    return '\n\n'.join(lines)

# Convert each chunk into text and store them in a list
text_chunks = [chunk_to_text(chunk) for chunk in chunks]

# To see the text format for each chunk
for i, text_chunk in enumerate(text_chunks):
    print(f"Chunk {i+1}:\n{text_chunk}\n")
```
