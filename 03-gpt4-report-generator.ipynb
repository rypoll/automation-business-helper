```python
import pandas as pd

# Load the Excel file
# file_path = "02-data/01a-link-list-manycam.csv"
# df = pd.read_csv(file_path, encoding='ISO-8859-1')
df = df_merged
# Function to count tokens
def count_tokens(text):
    return len(str(text).split())

# Add a new column for token count
df['token_count'] = df['why'].apply(count_tokens)

# Function to split dataframe into chunks based on token count
def split_dataframe(df, max_tokens=9000): # use 14k
    chunks = []  # List to store chunks of the dataframe
    
    while not df.empty:
        total_tokens = 0
        rows_to_take = []
        
        for index, row in df.iterrows():
            if total_tokens + row['token_count'] <= max_tokens:
                total_tokens += row['token_count']
                rows_to_take.append(index)
            else:
                break
        
        chunk = df.loc[rows_to_take]
        chunks.append(chunk)
        
        # Remove these rows from the original dataframe
        df = df.drop(rows_to_take)
    
    return chunks

# Call the function to split the dataframe
chunks = split_dataframe(df)

# To see the chunks and verify their token counts
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1} token count: {chunk['token_count'].sum()}")
```

```python
def chunk_to_text(chunk):
    """Convert a chunk (dataframe) to the desired text format."""
    # Use list comprehension to create a list of 'source: <source_value>\n<content_value>' for each row
    lines = [f"source: {row['source']}\nTitle: {row['content']} \nwhy: {row['why']}\nhours saved: {row['hours_saved']}" for idx, row in chunk.iterrows()]
    # Join all lines into a single text separated by two newlines
    return '\n\n'.join(lines)

# Convert each chunk into text and store them in a list
text_chunks = [chunk_to_text(chunk) for chunk in chunks]

# To see the text format for each chunk
for i, text_chunk in enumerate(text_chunks):
    print(f"Chunk {i+1}:\n{text_chunk}\n")
```

```python

import openai
from datetime import datetime
import os

# Define the folder and ensure it exists
answer_folder = 'reports'
if not os.path.exists(answer_folder):
    os.makedirs(answer_folder)


# Get the current date and time in the format ttddmmyy_hhmmss
current_date_time = datetime.now().strftime("%d%m%Y_%H%M%S")
file_name = os.path.join(answer_folder, f"{company_name}_{current_date_time}_gpt4_report.txt")


# Function to query the model and get a response
def get_response(text):
    content = '{prompt}: \n Business description: {business_outline} \n List of guides/automation tool use cases: {text}'.format(
        prompt=prompt, business_outline=business_outline, text=text.replace('"', '\''))
    messages = [{"role": "user", "content": content}]

    response = openai.ChatCompletion.create(
        # model="gpt-3.5-turbo-16k", # use 16k in the future, it's cheaper
        model="gpt-4-1106-preview",
        messages=messages
    )
    return response['choices'][0]['message']['content']

# Loop through the text_chunks
for idx, text in enumerate(text_chunks, 1):  # Start the index from 1 for more human-readable output
    success = False
    while not success:
        try:
            answer = get_response(text)
            print(answer)
            with open(file_name, 'a') as f:
                f.write(answer + '\n\n')
            success = True
            print(f"Processed {idx} of {len(text_chunks)} text_chunks.")
        except Exception as e:
            print(f"Error: {e}. Retrying...")
    break  # Add this line to break after the first iteration for testing purposes
```
