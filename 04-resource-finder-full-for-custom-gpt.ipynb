```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
import time

# Function to check if article text already exists in the .txt file
def is_text_in_file(text, file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        if text in file.read():
            return True
    return False

# Read the CSV file
df = pd.read_csv("02-data/04-combined-df-more-16.csv")

# Path to your .txt file
txt_file_path = '02-data/article_texts.txt'

# Your existing setup
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
chrome_options.add_argument(f'user-agent={USER_AGENT}')
chrome_options.add_argument("window-size=1920,1080")
driver_path = ChromeDriverManager().install()
chrome_service = Service(executable_path=driver_path)
chrome_service.service_log_path = 'NUL'
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

for index, row in df.iterrows():
    if pd.notna(row.get('article_text')):
        print(f"Skipping link {row['url']}, article text already present.")
        continue

    url = row['url']
    print(f"Processing link: {url}")

    retry_count = 0
    while retry_count < 3:
        try:
            driver.get(url)
            time.sleep(5)

            possible_containers = ["article", "main", "div.post-content", "div.article-content", "div.content"]
            article_text = ""

            for selector in possible_containers:
                elements = driver.find_elements(By.CSS_SELECTOR, selector)
                for element in elements:
                    if len(element.text) > len(article_text):
                        article_text = element.text

            if not is_text_in_file(article_text, txt_file_path):
                with open(txt_file_path, 'a', encoding='utf-8') as file:
                    file.write(f"Source: {row['source']}\nLink: {row['url']}\nArticle Text:\n{article_text}\n\n")
                print("Article text appended to .txt file.")

            df.at[index, 'article_text'] = article_text
            print("Article text added to DataFrame.")
            break
        except Exception as e:
            print(f"Error occurred: {e}. Retrying...")
            retry_count += 1
            time.sleep(3)

    if retry_count == 3:
        print(f"Failed to scrape {url} after 3 retries.")
        df.at[index, 'error'] = "Failed to scrape after 3 retries"

    df.to_csv("02-data/04-combined-df-more-16.csv", index=False)
    print(f"Progress saved to CSV for link: {url}")

driver.quit()
```

# Get links and titles from Zapier

```python
# Set up Chrome options
chrome_options = webdriver.ChromeOptions()
#chrome_options.add_argument('--headless')  # Run in headless mode if uncommented
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

# Set up the WebDriver
driver_path = ChromeDriverManager().install()
chrome_service = Service(executable_path=driver_path)
chrome_service.service_log_path = 'NUL'
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

# Load the existing data or initialize a new DataFrame
csv_file = '02-data/01b-zap-blog-titles.csv'
if os.path.exists(csv_file):
    print(f"Loading existing data from {csv_file}...")
    df_existing = pd.read_csv(csv_file)
else:
    df_existing = pd.DataFrame(columns=['type','url', 'content'])

# Create a set of URLs that have already been scraped
scraped_urls = set(df_existing['url'])

# List of URLs to scrape
base_url = 'https://zapier.com/blog/all-articles/zapier-tutorials/'
urls = [base_url] + [f'{base_url}{i}' for i in range(1, 15)]  # From 1 to 14

# Scrape each URL that hasn't been scraped yet
for url in urls:
    if url not in scraped_urls:
        print(f"Scraping {url}...")
        driver.get(url)
        time.sleep(10)  # Wait to ensure the page has loaded
        
        # Find all div elements with the specified data-testid attribute
        articles = WebDriverWait(driver, 30).until(
            EC.presence_of_all_elements_located((By.CSS_SELECTOR, 'div[data-testid="articleCard"]'))
        )
        
        # Extract the text from the <p> tags under each div
        for article in articles:
            p_texts = article.find_elements(By.CSS_SELECTOR, 'p')
            article_text = ' '.join([p.text for p in p_texts])
            # Append the new content to the DataFrame
            df_existing = pd.concat([df_existing, pd.DataFrame({'type': 'zapier','url': [url], 'content': [article_text]})], ignore_index=True)

        print(f"Scraped {len(articles)} articles from {url}")
    else:
        print(f"Content already scraped for {url}")

print("Finished scraping. Closing the browser...")
# Close the browser
driver.quit()

# Save the updated data to the CSV
print(f"Saving updated data to {csv_file}...")
df_existing.to_csv(csv_file, index=False)
print("Done!")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webedriver_manager.chrome import ChromeDriverManager
import pandas as pd
import os

# Set up Chrome options
chrome_options = webdriver.ChromeOptions()
# chrome_options.add_argument('--headless')  # Run in headless mode if uncommented
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')

print("Setting up the WebDriver...")
# Set up the WebDriver
driver_path = ChromeDriverManager().install()
chrome_service = Service(executable_path=driver_path)
chrome_service.service_log_path = 'NUL'
driver = webdriver.Chrome(service=chrome_service, options=chrome_options)

print("Loading the resource page...")
# Load the main resource page
driver.get("https://manychat.com/resources/how-to")

# Find all elements that contain the blog links and titles
print("Finding blog post elements...")
elements = driver.find_elements(By.CSS_SELECTOR, 'a.HowToTile_card__link__UTHmf')

# Initialize or load existing data
csv_file = '02-data/01c-manychat.csv'
if os.path.exists(csv_file):
    print(f"Loading existing data from {csv_file}...")
    df_existing = pd.read_csv(csv_file)
else:
    print("Initializing new DataFrame...")
    df_existing = pd.DataFrame(columns=['url', 'content'])

# Extract URLs and content, and add to DataFrame if not present
print("Scraping and processing data...")
for element in elements:
    href = element.get_attribute('href')
    full_url = f'{href}'
    content = element.find_element(By.CSS_SELECTOR, 'div.HowToTile_card__title__ZJt2u').text

    if full_url not in df_existing['url'].values:
        print(f"Adding new entry: {full_url}")
        df_existing = pd.concat([df_existing, pd.DataFrame({'url': [full_url], 'content': [content]})], ignore_index=True)
    else:
        print(f"URL already exists: {full_url}")

# Save the updated DataFrame to CSV
print(f"Saving updated data to {csv_file}...")
df_existing.to_csv(csv_file, index=False)

print("Scraping complete. Closing the browser...")
driver.quit()
```

```python
import pandas as pd

# Load the Excel file
file_path = "02-data/01a-link-list-manycam.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Function to count tokens
def count_tokens(text):
    return len(str(text).split())

# Add a new column for token count
df['token_count'] = df['content'].apply(count_tokens)

# Function to split dataframe into chunks based on token count
def split_dataframe(df, max_tokens=1500): # use 14k
    chunks = []  # List to store chunks of the dataframe
    
    while not df.empty:
        total_tokens = 0
        rows_to_take = []
        
        for index, row in df.iterrows():
            if total_tokens + row['token_count'] <= max_tokens:
                total_tokens += row['token_count']
                rows_to_take.append(index)
            else:
                break
        
        chunk = df.loc[rows_to_take]
        chunks.append(chunk)
        
        # Remove these rows from the original dataframe
        df = df.drop(rows_to_take)
    
    return chunks

# Call the function to split the dataframe
chunks = split_dataframe(df)

# To see the chunks and verify their token counts
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1} token count: {chunk['token_count'].sum()}")
```

```python
def chunk_to_text(chunk):
    """Convert a chunk (dataframe) to the desired text format."""
    # Use list comprehension to create a list of 'source: <source_value>\n<content_value>' for each row
    lines = [f"key: {idx}\nsource: {row['source']}\n{row['content']}" for idx, row in chunk.iterrows()]
    # Join all lines into a single text separated by two newlines
    return '\n\n'.join(lines)

# Convert each chunk into text and store them in a list
text_chunks = [chunk_to_text(chunk) for chunk in chunks]

# To see the text format for each chunk
for i, text_chunk in enumerate(text_chunks):
    print(f"Chunk {i+1}:\n{text_chunk}\n")
```

```python
import openai
from datetime import datetime
import os

# Define the folder and ensure it exists
answer_folder = 'answers'
if not os.path.exists(answer_folder):
    os.makedirs(answer_folder)


# Get the current date and time in the format ttddmmyy_hhmmss
current_date_time = datetime.now().strftime("%d%m%Y_%H%M%S")
file_name = os.path.join(answer_folder, f"{current_date_time}_answer.txt")


# Function to query the model and get a response
def get_response(text):
    content = '{prompt}: \n Business description: {business_outline} \n List of guides/automation tool use cases: {text}'.format(
        prompt=prompt, business_outline=business_outline, text=text.replace('"', '\''))
    messages = [{"role": "user", "content": content}]

    response = openai.ChatCompletion.create(
        # model="gpt-3.5-turbo-16k", # use 16k in the future, it's cheaper
        model="gpt-3.5-turbo",
        messages=messages
    )
    return response['choices'][0]['message']['content']

# Loop through the text_chunks
for idx, text in enumerate(text_chunks, 1):  # Start the index from 1 for more human-readable output
    success = False
    while not success:
        try:
            answer = get_response(text)
            print(answer)
            with open(file_name, 'a') as f:
                f.write(answer + '\n\n')
            success = True
            print(f"Processed {idx} of {len(text_chunks)} text_chunks.")
        except Exception as e:
            print(f"Error: {e}. Retrying...")
    break  # Add this line to break after the first iteration for testing purposes
```
